## I.- Problema 01


<img src="https://image.freepik.com/vector-gratis/varios-automoviles-dibujos-animados_23-2147613095.jpg" width="360" height="360" align="center"/>


EL conjunto de datos se denomina `Automobile_data.csv`, el cual contiene información tal como: compañia, precio, kilometraje, etc.

Lo primero es cargar el conjunto de datos y ver las primeras filas que lo componen:

```python
import pandas as pd
import os

# cargar datos
df = pd.read_csv(os.path.join("data","Automobile_data.csv"))
df.head()
```

El objetivo es tratar de obtener la mayor información posible de este conjunto de datos. Para cumplir este objetivo debe resolver las siguientes problemáticas:

1. Reemplazar los valores nulos (Nans) por "?"
2. Encuentra el nombre de la compañía de automóviles más cara
3. Imprimir todos los detalles de Toyota Cars
4. Cuente el total de automóviles por compañía
5. Encuentra el coche con el precio más alto por compañía
6. Encuentre el kilometraje promedio de cada compañía automotriz
7. Ordenar todos los autos por columna de precio

```python
# P1

print(list(df.columns))

for i in list(df.columns):
    print(sum(pd.isnull(df[i])))
```

```python
# Vemos que solo hay valores nulos en la columna 'price'

df['price'].fillna( method ='ffill', limit = 1, inplace = True) 
print(sum(pd.isnull(df['price'])))
```

```python
# P2
#print(df['company'].unique())
max = 0
masCaro = []
for i in list(df['company'].unique()):
    df2 = df.loc()[df['company'] == i]
    suma = sum(df2['price'])
    if(suma > max):
        masCaro = [i]
        max = suma
    elif(suma == max):
        masCaro.append(i)
print(masCaro, max)
```

La compañia mas cara es 'bmw' con un valor total de $163.280 [um]$.

```python
# P3
toyota = df.loc()[df['company'] == 'toyota']
toyota
```

```python
# P4
for i in list(df['company'].unique()):
    df2 = df.loc()[df['company'] == i]
    cant = df2.shape[0]
    print("La compañía", i, "tiene",cant,"automoviles.")
```

```python
# P5
for i in list(df['company'].unique()):
    df2 = df.loc()[df['company'] == i]
    precio = df2['price'].max()
    index = df2['price'].idxmax()
    print("El auto más caro de la compañía",i,"es el",index,"con un valor de",precio)
```

```python
# P6
for i in list(df['company'].unique()):
    df2 = df.loc()[df['company'] == i]
    prom = df2['average-mileage'].mean()
    print("El promedio de kilometraje de la compañía",i,"es",round(prom))
```

```python
# P7

autos = df
autos.sort_values(by = 'price', inplace=True)
autos
```

## II.- Problema 02


Siguiendo la temática de los automóviles, resuelva los siguientes problemas:

#### a) Subproblema 01
A partir de los siguientes diccionarios:

```python
germanCars = {'Company': ['Ford', 'Mercedes', 'BMV', 'Audi'],
              'Price': [23845, 171995, 135925 , 71400]}
japaneseCars = {'Company': ['Toyota', 'Honda', 'Nissan', 'Mitsubishi '], 
                'Price': [29995, 23600, 61500 , 58900]}
```

* Cree dos dataframes (**carsDf1** y **carsDf2**) según corresponda.
* Concatene ambos dataframes (**carsDf**) y añada una llave ["Germany", "Japan"], según corresponda.

```python
carsDf1 = pd.DataFrame(germanCars)
carsDf1
```

```python
carsDf2 = pd.DataFrame(japaneseCars)
carsDf2
```

```python
carsDf = pd.concat([carsDf1, carsDf2],keys=['German','Japanese'])
carsDf
```

A partir de los siguientes diccionarios:
