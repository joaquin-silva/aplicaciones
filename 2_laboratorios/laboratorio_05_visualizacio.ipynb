El objetivo es tratar de obtener la mayor información posible de este conjunto de datos. Para cumplir este objetivo debe resolver las siguientes problemáticas:

1. Lea el "total_profit" de todos los meses y muéstrelo usando un gráfico lineal
2. Lea todos los datos de ventas de productos y muéstrelos utilizando un gráfico multilínea.
3. Lea los datos de ventas de "toothpaste" de cada mes y muéstrelos usando un diagrama de dispersión

**Observación.-** Puedes ocupar las librerías de Matplolib o Seaborn.

```python
# Librerías
import matplotlib.pyplot as plt
import seaborn as sns
```

## Pregunta 1

```python
sns.set_style("darkgrid")
plt.plot(df['month_number'],df['total_profit'],c='red')
plt.xlabel('Month number')
plt.ylabel('Total profit')
plt.show()
```

## Pregunta 2

```python
sns.set_style("darkgrid")
plt.figure(figsize = (8,6))
plt.plot(df['month_number'], df['facecream'])
plt.plot(df['month_number'],df['facewash'])
plt.plot(df['month_number'],df['toothpaste'])
plt.plot(df['month_number'],df['bathingsoap'])
plt.plot(df['month_number'],df['shampoo'])
plt.plot(df['month_number'],df['moisturizer'])
plt.xlabel('Month number')
plt.legend(['Face cream','Face wash','Thooth paste','Shampoo','Moisturizer'],loc='lower center', bbox_to_anchor=(0.5, 1.00), shadow=True, ncol=5)
plt.show()
```

## Pregunta 3

```python
sns.set_style("darkgrid")
plt.plot(df['month_number'],df['toothpaste'],'.')
plt.xlabel('Month number')
plt.ylabel('Toothpaste')
plt.show()
```

## Pregunta 4

```python
import numpy as np
labels = df['month_number']
face_cream = df['facecream']
face_wash = df['facewash']

x = np.arange(len(labels))  # the label locations
width = 0.35  # the width of the bars

fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, face_cream, width, label='Face Cream')
rects2 = ax.bar(x + width/2, face_wash, width, label='Face Wash')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Ventas')
ax.set_title('Ventas por grupo')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()
```

## Pregunta 5

```python
bathingsoap = df['bathingsoap']

fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, bathingsoap, width, label='Bathing soap')

ax.set_ylabel('Ventas')
ax.set_xlabel('Meses')
ax.set_title('Ventas Bathing soap')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()

# Guardar imagen
plt.savefig("grafico.jpg")
```

# Pregunta 6

```python
total_facecream = sum(df['facecream'])
total_facewash = sum(df['facewash'])
total_toothpaste = sum(df['toothpaste'])
total_bathingsoap = sum(df['bathingsoap'])
total_shampoo = sum(df['shampoo'])
total_moisturizer = sum(df['moisturizer'])

# Datos a graficar
labels = 'Face cream', 'Face wash', 'Toothpaste', 'Bathing soap', 'Shampoo', 'Moisturizer'
sizes = [total_facecream, total_facewash, total_toothpaste, total_bathingsoap, total_shampoo,total_moisturizer]

# Gráfico
plt.figure(figsize=(6,6))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', shadow=True, startangle=140)
plt.axis('equal')
plt.show()
```
